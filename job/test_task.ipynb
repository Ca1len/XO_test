```python
leads_time_x2 = true_leads_11_month_half.merge(transactions, how='inner', on='l_client_id')
new_leads_time_x2 = new_leads.merge(transactions, how='inner', on='l_client_id')
```

```python
leads_time_x2.rename(columns={'created_at_x': 'created_at', 'created_at_y': 'trans_created_at'}, inplace=True)
new_leads_time_x2.rename(columns={'created_at_x': 'created_at', 'created_at_y': 'trans_created_at'}, inplace=True)
```

```python
leads_time_x2.head()
```

```python
buyers_and_trans = leads_time_x2.loc[((leads_time_x2.trans_created_at - leads_time_x2.created_at).dt.days < 7) &
                          ((leads_time_x2.trans_created_at - leads_time_x2.created_at).dt.days > 0)]
new_buyers_and_trans = new_leads_time_x2.loc[((new_leads_time_x2.trans_created_at - new_leads_time_x2.created_at).dt.days < 7) &
                          ((new_leads_time_x2.trans_created_at - new_leads_time_x2.created_at).dt.days > 0)]
```

```python
buyers = buyers_and_trans.groupby('l_client_id').head(1)
new_buyers = new_buyers_and_trans.groupby('l_client_id').head(1)
```

```python
table_buyers = pd.pivot_table(buyers, values='lead_id',
                      index=['d_manager', 'd_club', 'd_utm_source'],
                      aggfunc={
                          'lead_id': 'count'
                      }).rename(columns={'lead_id': 'buyers'}).reset_index()
table_new_buyers = pd.pivot_table(new_buyers, values='lead_id',
                      index=['d_manager', 'd_club', 'd_utm_source'],
                      aggfunc={
                          'lead_id': 'count'
                      }).rename(columns={'lead_id': 'new_buyers'}).reset_index()
```
